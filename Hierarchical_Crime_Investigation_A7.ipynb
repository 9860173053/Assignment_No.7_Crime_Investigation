# A.Hierarchical Clustering

### 1. Import Necessary Libraries

```python
import pandas as pd
import numpy as Np
from matplotlib import pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

# 2.Import data

```python
Data=pd.read_csv('crime_data.csv')
Data
```

###  First Step= Normalized the data

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(Data.iloc[ : ,1: ])
df_norm
```

```python
#Let’s first draw the dendrogram to help us decide the number of clusters for this particular problem:
```

### Step 2: Creating Dendogram

```python
dendogram=sch.dendrogram(sch.linkage(df_norm,method='ward'))
```

### Step 3- Creating Clusters

```python
hc=AgglomerativeClustering( n_clusters=5, affinity='euclidean',linkage='ward')
```

```python
y=pd.DataFrame(hc.fit_predict(df_norm),columns=['clusters'])
y
```

```python
# Adding clusters to dataset
Data['clusterid']=hc.labels_
Data
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(Data['clusterid'],Data['Murder'], c=hc.labels_) 
```
